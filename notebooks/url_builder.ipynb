# crop_mode-pad_resize

## 1. One-line description

`crop_mode-pad_resize` resizes the image to fit within the requested width and height **without cropping**, preserving aspect ratio by adding padding to reach exact dimensions.

---

## 2. Limitations

* Requires **both `w` and `h`** to be meaningful.
* Never crops image content.
* Output may contain padding (transparent or background color).
* Aspect ratio is always preserved.
* Padding placement depends on `fo` (default: center).
* Not suitable if you want to remove any part of the image.

# One-line descriptions

**force (`force`)**
Resizes the image to the exact requested width and height **without preserving aspect ratio**, squeezing or stretching the image if needed.

**at_max_enlarge (`at_max_enlarge`)**
Resizes the image while preserving aspect ratio so that it fits **within the requested dimensions**, allowing enlargement beyond the original size.

**at_least (`at_least`)**
Resizes the image while preserving aspect ratio so that it is **at least as large as the requested dimensions**, possibly exceeding one dimension.

**maintain_ratio (`maintain_ratio`)**
Resizes the image to the exact requested width and height while preserving aspect ratio by **cropping parts of the image** (default crop behavior).

---

# Limitations

## force (`force`)

* Requires **both `w` and `h`**
* Aspect ratio is **not preserved**
* No cropping occurs
* Image may appear distorted
* Focus (`fo`) has no effect
* Not suitable for photos where proportions matter

---

## at_max_enlarge (`at_max_enlarge`)

* Requires **both `w` and `h`**
* Aspect ratio is always preserved
* Output dimensions may be **smaller than requested**
* Enlarges images beyond original size
* No cropping occurs
* Focus (`fo`) has no effect

---

## at_least (`at_least`)

* Requires **both `w` and `h`**
* Aspect ratio is preserved
* One dimension may be **larger than requested**
* No cropping occurs
* Focus (`fo`) has no effect
* Image may overflow container in one dimension

---

## maintain_ratio (`maintain_ratio`)

* Requires **both `w` and `h`**
* Cropping always occurs if aspect ratios differ
* Aspect ratio is preserved
* Default focus is center
* Sensitive to `fo` (auto, face, object, custom)
* Default crop mode if no `c` or `cm` is specified

---

# One-line description

**zoom (`z`)**
Controls how much to zoom in or zoom out around the focused area during face- or object-based cropping.

---

# Limitations

* Works **only with focus-based cropping**

  * `fo-face`
  * `fo-<object>`
* Has **no effect** without a valid `fo` value
* Does **not** work with:

  * `c-force`
  * `c-at_max`
  * `c-at_least`
  * `c-at_max_enlarge`
  * `cm-pad_resize`
* Default value is `z-1`
* Valid values:

  * `0 < z < 1` → zoom out (include more background)
  * `z > 1` → zoom in (exclude background)
* Accuracy depends on quality of object / face detection
* Behavior differs based on crop mode:

  * Default `c-maintain_ratio`
  * `cm-extract`
  * Tightly cropped object

# One-line description

**zoom (`z`)**
Controls how much to zoom in or zoom out around the focused area during face- or object-based cropping.

---

# Limitations

* Works **only with focus-based cropping**

  * `fo-face`
  * `fo-<object>`
* Has **no effect** without a valid `fo` value
* Does **not** work with:

  * `c-force`
  * `c-at_max`
  * `c-at_least`
  * `c-at_max_enlarge`
  * `cm-pad_resize`
* Default value is `z-1`
* Valid values:

  * `0 < z < 1` → zoom out (include more background)
  * `z > 1` → zoom in (exclude background)
* Accuracy depends on quality of object / face detection
* Behavior differs based on crop mode:

  * Default `c-maintain_ratio`
  * `cm-extract`
  * Tightly cropped object

# One-line descriptions

**x**
Specifies the horizontal position of the **top-left corner** of the extracted area from the original image.

**y**
Specifies the vertical position of the **top-left corner** of the extracted area from the original image.

**x_center**
Specifies the horizontal position of the **center point** of the extracted area from the original image.

**y_center**
Specifies the vertical position of the **center point** of the extracted area from the original image.

---

# Limitations

## x

* Works **only with `cm-extract`**
* Requires **both `w` and `h`**
* Value is relative to the original image’s top-left corner
* Crop fails if extracted area ex_centereeds image bounds
* Affected by image auto-rotation metadata
* Cannot be used with resize-only crop modes

---

## y

* Works **only with `cm-extract`**
* Requires **both `w` and `h`**
* Value is relative to the original image’s top-left corner
* Crop fails if extracted area ex_centereeds image bounds
* Affected by image auto-rotation metadata
* Cannot be used with resize-only crop modes

---

## x_center

* Works **only with `cm-extract`**
* Requires **both `w` and `h`**
* Defines center of extraction, not the edge
* Crop fails if centered area ex_centereeds image bounds
* More error-prone than `x,y`
* Affected by image auto-rotation metadata

---

## y_center

* Works **only with `cm-extract`**
* Requires **both `w` and `h`**
* Defines center of extraction, not the edge
* Crop fails if centered area ex_centereeds image bounds
* More error-prone than `x,y`
* Affected by image auto-rotation metadata

# One-line description

**dpr (Device Pixel Ratio)**
Specifies the device pixel ratio used to scale the output image dimensions for high-DPR displays like Retina or high-end mobile screens.

---

# Limitations

* Can be used **only when `w` or `h` is specified**
* Valid range: **`0.1` to `5`**
* If calculated output width or height:

  * Is **less than 1px**, or
  * Is **greater than 5000px**
    → DPR is ignored and base dimensions are used
* Does **not** change aspect ratio
* Does **not** work without dimensions
* `dpr-auto` depends on client hints (`Sec-CH-DPR`)
* Ineffective if client hints are not sent by the browser

---
